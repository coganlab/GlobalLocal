```python
import pandas as pd
import os

subjects = ['D0057', 'D0059', 'D0063', 'D0065', 'D0069', 'D0071', 'D0077', 
            'D0090', 'D0094', 'D0100', 'D0102', 'D0103', 'D0107A', 'D0110', 
            'D0116', 'D0117', 'D0121', 'D0130']

clean_root = r'C:\Users\jz421\Box\CoganLab\BIDS-1.1_GlobalLocal\BIDS\derivatives\clean'

eeg_names = ['T5', 'T6', 'FZ', 'CZ', 'PZ', 'FP1', 'FP2', 'F3', 'F4', 
             'C3', 'C4', 'P3', 'P4', 'O1', 'O2', '02', 'F7', 'F8', 'T3', 'T4']

for sub in subjects:
    tsv = os.path.join(clean_root, f'sub-{sub}', 'ieeg',
                       f'sub-{sub}_task-GlobalLocal_acq-01_run-01_desc-clean_channels.tsv')
    if not os.path.exists(tsv):
        print(f"{sub}: channels.tsv not found")
        continue
    df = pd.read_csv(tsv, sep='\t')
    found = [ch for ch in eeg_names if ch in df['name'].values]
    if found:
        print(f"{sub}: has EEG channels {found}")
```
