```python
def get_baseline(inst: mne.io.BaseRaw, base_times_list: list[tuple[float, float]]):
    combined_trials = None
    
    inst = inst.copy()
    inst.load_data()
    ch_type = inst.get_channel_types(only_data_chs=True)[0]
    inst.set_eeg_reference(ref_channels="average", ch_type=ch_type)
    
    for base_times in base_times_list:
        adjusted_base_times = [base_times[0] - 0.5, base_times[1] + 0.5]
        trials = trial_ieeg(inst, "experimentStart", adjusted_base_times, preload=True)
        
        if trials is None or len(trials) == 0:
            print(f"Warning: No trials found for base_times {base_times}. Skipping...")
            continue
        
        # outliers_to_nan(trials, outliers=10)
        if combined_trials is None:
            combined_trials = trials
        else:
            # Check for consistency before concatenating
            if not all(ch1 == ch2 for ch1, ch2 in zip(combined_trials.ch_names, trials.ch_names)):
                raise ValueError("Mismatched channel names between trials. Ensure all trials have the same channels.")
            # Concatenate the trials
            combined_trials = mne.concatenate_epochs([combined_trials, trials])
    
    if combined_trials is None:
        raise ValueError("No valid trials found for the provided base_times_list.")
    
    HG_base = gamma.extract(combined_trials, copy=False, n_jobs=1)
    broadband_base = gamma.extract(combined_trials, copy=False, n_jobs=1, passband=(0,1000))
    crop_pad(HG_base, "0.5s")
    crop_pad(broadband_base, "0.5s")
    
    del inst
    return HG_base, broadband_base


base_times = [(1,101)]
HG_base, broadband_base = get_baseline(filt, base_times)
```
